# Objective

- Connect mongodb using pymongo
- run some basic queries: find(), find_one() 

```python
from pymongo.mongo_client import MongoClient
from pymongo.server_api import ServerApi ## for mongodb connection
import pandas as pd 
import os

from dotenv import load_dotenv # for secure storage  key and secret
load_dotenv() 
```

```python
user = os.getenv("MONGO_USER")
password = os.getenv("MONGO_PASSWORD")
string = "mongodb+srv://{p1}:{p2}@cluster0.nuy0m.mongodb.net/test?retryWrites=true&w=majority"
```

```python
client = MongoClient(string.format(p1=user, p2=password), server_api=ServerApi('1'))
```

```python
# Choose a database
db = client.sample_airbnb
```

```python
db.list_collection_names()
```

```python
# Choose a collection within the database (similar to table in SQL)
collection = db.listingsAndReviews
type(collection)
```

```python
# total number of documents 
collection.count_documents({})
```

```python
# show all index of the collection, similar to columns in Relational db
'''
Returns a dictionary where the keys are index names (as returned by create_index()) and the values are dictionaries containing information about each index. 
The dictionary is guaranteed to contain at least a single key, "key" which is a list of (key, direction) pairs specifying the index (as passed to create_index())
'''
collection.index_information()
```

```python
# find place's name where maximum_nights: 30
max_30 = collection.find_one({'maximum_nights': '30'})
```

```python
# Get id and name from max_30

[max_30.get(key) for key in ['_id', 'name']]
```

```python
# Can also be done using pandas

m = pd.Series(max_30)
m[['_id', 'name']]
```

```python
collection.find_one({"_id": "1206363"})
```

```python
# Count the documents where bedrooms: 1

collection.count_documents({'bedrooms': 1})
```

```python
# Greater or less than 
# find document where beds > 3 AND bedrooms > 5

collection.find_one({'beds': {"$gt": 3},'bedrooms': {"$gt": 5}})
```
